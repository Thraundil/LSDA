```python
import os
import json
import pandas as pd

from tqdm import tqdm

os.chdir('D:\\02. LSDA\\github\\02_code')
dataset = '../01_input_data/train.json'

with open(dataset, 'r') as f:
    data = json.load(f)
images = pd.DataFrame(data['images'])
annotations = pd.DataFrame(data['annotations'])
del data
```

```python
#For each annotation get the frequency... Then get the first 100 keywords for that image..
anno_tuples = []
for name,row in tqdm(annotations.set_index('imageId').iterrows(),total = len(annotations)):
  for label in row['labelId']:
    anno_tuples += [(name,label)]
annotations = pd.DataFrame(anno_tuples)
annotations.columns = ['imageId','labelId']

first_100_annos = pd.DataFrame(columns = annotations.columns)
for name, group in annotations.groupby('labelId'):
  first_100_annos = first_100_annos.append(group.iloc[:100])
```

```python
 
from IPython.display import Image
from IPython.display import display

df = pd.merge(first_100_annos,images)
import matplotlib.pyplot as plt
for name,row in df[df['labelId']=='34'].iterrows():
  display(Image(url=row['url'],height=200,width=200))
```

```python
stack = []
for imageId in first_100_annos[first_100_annos['labelId'] == '33']['imageId'].values:
  if imageId in titles.keys():
    stack += keywords[imageId]
pd.Series(stack).value_counts()
```
